```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create the sampling domain
x = [0, 1]
y = [0, 2]
# create lines that defines the triangles
X = np.linspace(0, 1, 100)
Y = np.linspace(0, 2, 100)
y1 = lambda y: 4*y
y2 = lambda y: -4*y + 4
```

```python
plt.figure(figsize=(5, 5))
plt.plot(X, y1(X), color='b')
plt.plot(X, y2(X), color='b')
plt.xlim(x)
plt.ylim(y)
plt.show()
```

```python
# draw the sampling domain

nSamples = 2000

plt.figure(figsize=(5, 5))
plt.plot(X, y1(X), color='b')
plt.plot(X, y2(X), color='b')
# plot a sample
count = 0
for _ in range(nSamples):
    # draw random samples from a uniform distribution
    # samples from the y axis
    y = [0, 2] 
    r1 = random.uniform(x[0], x[1])
    # samples from the x axis
    x = [0, 1]
    r2 = random.uniform(y[0], y[1])
    
    # draw samples on the graph
    if (r1 < np.mean(x) and r2 > y1(r1)) or (r1 > np.mean(x) and r2 > y2(r1)):
        plt.scatter(r1, r2, c='k')
    else:
        plt.scatter(r1, r2, c='r')
        # update counter when inside the triangle
        count += 1

# calculate the probability of being within the triangle
prob = count / nSamples
# the area of the triangle is the probability of being inside multiplied with the sampling domain
areaOfTriangle = prob * (y[1] - y[0]) *(x[1] - x[0])

plt.title("Area of red dots: {}".format(areaOfTriangle))
plt.xlim(x)
plt.ylim(y)
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
# calculate the area of a comlicated double integral

# define the sampling domain

y = [0, 2]
x = [0, 2]

V = (y[1] - y[0]) * (x[1] - x[0])
```

```python
# define the complicated function to sample 
def complicatedFunction(x, y):
    f1 = np.cos(x + y)
    f2 = np.sin( np.sqrt(x) * y )
    f3 = 1 + x**2 + y**2

    return (f1*f2) / f3
```

```python
# define the experiment

nSamples = 10000

sumComplicatedFunction = 0
for _ in range(nSamples):

    # draw samples from a uniform distribution within the domain 
    yr = random.uniform(0, 2)
    xr = random.uniform(0, 2)

    sumComplicatedFunction += complicatedFunction(xr, yr)
I = V * (1 / nSamples) * sumComplicatedFunction
print("The integration of the function in the domain: {}".format(I))

```
